```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python

# 1) Load the YOLOv8 segmentation model (e.g., 'yolov8s-seg.pt')
model = YOLO("yolov8s-seg.pt")

results = model.predict(sample_1)

# 3) Grab the first (and only) result if you're processing a single image
res = results[0]

# 4) res.plot() returns a NumPy array (annotated image) in BGR format
annotated_img = res.plot()

# 5) (Optional) Read the original image to show it side by side
original_bgr = cv2.imread(sample_1)
original_rgb = cv2.cvtColor(original_bgr, cv2.COLOR_BGR2RGB)

# 6) Convert the annotated image from BGR to RGB for display with matplotlib
annotated_rgb = annotated_img[..., ::-1]

# 7) Plot side by side using matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.imshow(original_rgb)
ax1.set_title("Original Image")
ax1.axis("off")

ax2.imshow(annotated_rgb)
ax2.set_title("YOLOv8 Segmentation")
ax2.axis("off")

plt.tight_layout()
plt.show()
```
